```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.agents import Tool
```

```python
#Load API key
load_dotenv(dotenv_path=".env")
openai_api_key=os.getenv("OPENAI_API_KEY")
```

```python
#Initialize LLM
llm=ChatOpenAI(model="gpt-4o-mini", temperature=0)
#Create Prompt Template
prompt=PromptTemplate.from_template("Answer clearly: {question}")
```

```python
#Build the LLMChain
qa_chain= LLMChain(llm=llm, prompt=prompt)

#Wrap LLMChain as a Tool
qa_tool=Tool(
    name="Simple QA",
    func=qa_chain.run,
    description="A basic LLM chain that answers clear questions"
)
```

```python
# Use the Tool directly
question="What is LangGraph in LangChain?"
answer=qa_tool.run(question)

#Output
print("\nUser Question:", question)
print("\nTool Answer:", answer)
```
