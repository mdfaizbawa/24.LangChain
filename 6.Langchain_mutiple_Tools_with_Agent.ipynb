```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.agents import Tool, initialize_agent, AgentType
from langchain.tools import TavilySearchResults
```

```python
#Load API key
load_dotenv(dotenv_path=".env")
openai_api_key=os.getenv("OPENAI_API_KEY")
tavily_api_key = os.getenv("TAVILY_API_KEY")

#Initialize LLM
llm= ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
#Tool 1: Simple QA Tool

qa_prompt=PromptTemplate.from_template("Answer clearly: {question}")
qa_chain=LLMChain(llm=llm, prompt=qa_prompt)
qa_tool=Tool(
    name="Simple QA",
    func=qa_chain.run,
    description="Answer factual questions clearly"
)

#Tool 2: Summarizer Tool

summary_prompt= PromptTemplate.from_template("Summarize the following text:\n\n{text}")
summary_chain=LLMChain(llm=llm, prompt=summary_prompt)
summary_tool=Tool(
    name="Summarizer",
    func=summary_chain.run,
    description="Summarizes input text"
)

#Tool 3: Web Search (Tavily)
search_tool=Tool(
    name="Web Search",
    func=TavilySearchResults(max_results=3).run,
    description="Search the internet for current information"
)

#Wrap all tools in an agent
tools=[qa_tool, summary_tool, search_tool]

agent_executor=initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

queries=["What is LangGraph in LangChain?",
       "LangGraph is a library built on top of LangChain that helps developers create stateful, multi-step agents as graphs. Each node represents a step like calling an LLM or a tool. It's ideal for advanced AI workflows.",
       "Latest updates on GPT-4o by OpenAI."
      
]

#Run tools manually
for query in queries:
    print("\n User Query:", query)
    response=agent_executor.run(query)
    print("\n Agent Response:", response)
```
